```python
import os

def save_results(content: str, directory="./crawl4ai_output/"):
    os.makedirs(directory, exist_ok=True)

    file_path = os.path.join(directory, "scraped_content.md")  

    with open(file_path, "w", encoding="utf-8") as f:
        f.write(content)
```

```python
from crawl4ai import *

async with AsyncWebCrawler() as crawler:
    result = await crawler.arun(
        url="https://www.vensure.com/peo-services",
    )
    print(result)

if result:
    save_results(result.markdown)
```

```python
import asyncio
import nest_asyncio
import os

from base64 import b64decode
from crawl4ai import AsyncWebCrawler, CacheMode

nest_asyncio.apply()

async def main():
    async with AsyncWebCrawler() as crawler:
        result = await crawler.arun(
            url="https://www.vensure.com/peo-services/",
            cache_mode=CacheMode.BYPASS,
            pdf=True,
            screenshot=True
        )

        if result.success:
            if result.screenshot:
                with open("./crawl4ai_output/vensure_screenshot.png", "wb") as f:
                    f.write(b64decode(result.screenshot))

            if result.pdf:
                with open("./crawl4ai_output/vensure_page.pdf", "wb") as f:
                    f.write(result.pdf)

            print("[OK] PDF & Screenshot captured.")
        else:
            print("[ERROR]", result.error_message)


asyncio.run(main())
```
